```python
#Importing the necessary Python libraries
import pandas as pd
import numpy as np
import matplotlib as plt
%matplotlib inline
import seaborn as sns
```

```python
#Loading and Reading the dataset

data = pd.read_csv('pokemon.csv', index_col = 'name')
print(data.shape)
```

```python
data = data[['against_bug', 'against_dark', 'against_dragon',
       'against_electric', 'against_fairy', 'against_fight', 'against_fire',
       'against_flying', 'against_ghost', 'against_grass', 'against_ground',
       'against_ice', 'against_normal', 'against_poison', 'against_psychic',
       'against_rock', 'against_steel', 'against_water']]
```

```python
data['resistant'] = 0
data['weaknesses'] = 0
index = list(range(0,18))
print(index)
def weak(row):
    for i in index:
        if row[i] > 1: 
            row.weaknesses += 1

        else:
            row.weaknesses += 0
    return row.weaknesses        
data.weaknesses = data.apply(weak, axis='columns')

def resistant(row):
    for i in index:
        if row[i] > 1: 
            row.resistant += 1

        else:
            row.resistant += 0
    return row.resistant       
data.resistant = data.apply(resistant, axis='columns')
data.head(70)
```

```python
# Finding the heaviest pokemon of a given type in a given generation using multi-index grouping and a pyhton function.

grouped = data.groupby(['generation', 'type1'])

#for name,group in grouped:
#while name[0] == 1:
gen = input('Please enter a gen name')
types = input('Please enter a pokemon type')
def fastest_pokemon(gen,types):
    pokemon = grouped.get_group((int(gen),types))
    speed = pokemon['weight_kg'].idxmax()
    print(f'The heaviest {types} type pokemon in generation {gen} is {speed}')
    
fastest_pokemon(gen,types)
```
